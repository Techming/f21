```python
import pandas as pd
```

```python
df = pd.read_csv("waves.csv")
df.head()
```

```python
df = df[df["Wave Height"] >= 0]
```

```python
df.plot.scatter(x="Wave Period", y="Wave Height")
```

```python
beaches = sorted(set(df["Beach Name"]))
beaches
```

```python
for beach in beaches:
    df[df["Beach Name"] == beach].plot.scatter(x="Wave Period", y="Wave Height")
```
